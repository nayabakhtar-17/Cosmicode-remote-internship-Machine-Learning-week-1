```python
# Importing Required Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.feature_selection import mutual_info_regression
from scipy import stats

%matplotlib inline
```

```python
# Task 1 : Linear Regression from Scratch using NumPy

# Sample Data
X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

# Linear Regression Function
def linear_regression(X, y):
    n = len(X)
    m = (n * np.sum(X*y) - np.sum(X) * np.sum(y)) / (n * np.sum(X**2) - (np.sum(X))**2)
    b = (np.sum(y) - m * np.sum(X)) / n
    return m, b

# Train Model
m, b = linear_regression(X, y)
y_pred = m * X + b

# Plot Regression Line
plt.scatter(X, y, label="Data Points")
plt.plot(X, y_pred, color='red', label="Regression Line")
plt.title("Linear Regression using NumPy")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```

```python
# Task 2: Data Cleaning with Seaborn’s tips Dataset

# Load a real-world dataset
df = sns.load_dataset('tips')
df.head()
# Check for Missing Values
print(df.isnull().sum())

# Handle Missing Values (if any)
df.fillna(df.mean(numeric_only=True), inplace=True)

# Detect & Remove Outliers using Z-score
z_scores = np.abs(stats.zscore(df.select_dtypes(include='number')))
df_clean = df[(z_scores < 3).all(axis=1)]

# Normalize the Data
scaler = StandardScaler()
df_scaled = df_clean.copy()
df_scaled[df_clean.select_dtypes(include='number').columns] = scaler.fit_transform(
    df_clean.select_dtypes(include='number')
)

df_scaled.head()
```

```python
# Task 3: Feature Selection using Mutual Information

# Correlation Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(df_scaled.select_dtypes(include=np.number).corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()

# Mutual Information (predicting 'tip')
X = df_scaled.drop('tip', axis=1)
y = df_scaled['tip']

# Use regression version since 'tip' is numeric
# For mutual information, we need to encode categorical variables
X_encoded = X.copy()
for col in X_encoded.select_dtypes(include=['category', 'object']).columns:
    X_encoded[col], _ = X_encoded[col].factorize()

mi = mutual_info_regression(X_encoded, y)
mi_series = pd.Series(mi, index=X.columns)

# Plot Feature Importances
mi_series.sort_values(ascending=False).plot.bar(color='skyblue')
plt.title("Mutual Information Feature Importance")
plt.ylabel("Score")
plt.grid(True)
plt.show()
```

```python
# Task 4: Exploratory Data Analysis (EDA)

# Univariate: Histograms
df_clean.hist(figsize=(12, 8), bins=20)
plt.tight_layout()
plt.show()
# Box Plots for Numerical Features
for col in df_clean.select_dtypes(include='number').columns:
    plt.figure(figsize=(6, 1.5))
    sns.boxplot(data=df_clean, x=col, color='lightgreen')
    plt.title(f"Boxplot - {col}")
    plt.show()
# Pair Plot (can be heavy, run only if needed)
sns.pairplot(df_clean)
plt.show()
```

```python
# Task 5: PCA – Principal Component Analysis

# PCA on Scaled Data
X_pca = df_scaled.select_dtypes(include='number')

pca = PCA(n_components=2)
pca_result = pca.fit_transform(X_pca)

# Visualize in 2D
plt.scatter(pca_result[:, 0], pca_result[:, 1], c=df_scaled['tip'], cmap='viridis')
plt.title('PCA: 2D Scatter Plot')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Tip Amount (scaled)')
plt.grid()
plt.show()
```
